# **Checking Data Accuracy and Timeline Consistency**

```python
%%sql
SELECT c.date_last_updated,
       c.age,
       c.customer_identifier,
       COUNT(DISTINCT(t.account_number)) AS num_accounts,
       COUNT(t.transaction_description) as num_of_trans,
       i.income_group_desc
FROM customer c
JOIN transactiondata t 
ON t.customer_identifier = c.customer_identifier
JOIN income_group i
ON i.income_group_code = c.income_group_code
GROUP BY c.date_last_updated,
         c.customer_identifier,
         c.age,
         i.income_group_desc
ORDER BY c.date_last_updated
LIMIT 20;
```

The dataset shows inconsistencies, particularly with dates (e.g., 1922) that precede the bank's founding in 1991, suggesting potential data errors or placeholder values. While customer data (age, transactions) appears realistic, it's disconnected from the timeline. Financial details like income groups also seem modern for the implied period. Recommend verifying date accuracy, clarifying dataset intent, and validating data fields before proceeding with analysis.

# INACTIVE CUSTOMERS

```python
%%sql
SELECT 
    COUNT(DISTINCT c.customer_identifier) AS customer_count,
    CASE WHEN count(t.transaction_description) < 0 THEN 1 ELSE 0 END AS num_of_trans
FROM 
    customer c
JOIN 
    transactiondata t 
ON 
    t.customer_identifier = c.customer_identifier;

```

```python
%%sql
SELECT 
    COUNT(DISTINCT customer_identifier) AS Inactive_Customers,
    CASE 
        WHEN COUNT(transaction_description) = 0 THEN 1 
        ELSE 0 
    END AS num_of_trans_zero
FROM transactiondata
;

```

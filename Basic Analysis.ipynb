### Number of Customers

```python
%%sql
SELECT COUNT(DISTINCT(customer_identifier)) FROM customer
```

```python
%%sql
SELECT DISTINCT(date_last_updated) from customer
```

```python
%%sql
SELECT DISTINCT(sex_code) from customer
```

```python
%%sql
SELECT DISTINCT(income_group_code),COUNT(customer_identifier) from customer
GROUP BY income_group_code
ORDER BY income_group_code
```

```python
%%sql
SELECT DISTINCT(number_of_accounts), COUNT(customer_identifier) AS Number_of_Customers from customer
GROUP BY number_of_accounts
ORDER BY Number_of_Customers
```

```python
%%sql
SELECT DISTINCT(occupational_status_code), COUNT(customer_identifier) AS Number_of_Customers from customer
GROUP BY occupational_status_code
ORDER BY occupational_status_code
```

```python
%%sql
SELECT DISTINCT(age), COUNT(customer_identifier) AS Number_of_Customers from customer
GROUP BY age
ORDER BY age
```
